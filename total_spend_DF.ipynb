```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as func
spark = SparkSession.builder.appName("total_spend").getOrCreate()
```

```python
df1 = spark.read.csv("file:///SparkCourse/customer-orders.csv", schema="customerID int, itemID int, amount_spend float")
df1.printSchema()
df2 = df1.select("customerID","amount_spend").groupBy("customerID").agg(func.round(func.sum("amount_spend"),2))
results = df2.collect()
for result in results:
    print(f"{result[0]} \t {result[1]}")
```
