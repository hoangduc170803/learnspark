```python
import re
from pyspark import SparkConf,SparkContext
conf = SparkConf().setMaster("local").setAppName("WordCount")
sc = SparkContext(conf = conf)
```

```python
def normalizeWords(text):
    # tìm kiếm tất cả các "từ" trong chuỗi text với r'\W+' , re.UNICODE cho phép nhận diện kí tự UNICODE
    return re.compile(r'\W+', re.UNICODE).split(text.lower())
```

```python
input = sc.textFile("file:///SparkCourse/Book")
words = input.flatMap(normalizeWords)
wordCounts = words.map(lambda x: (x,1)).reduceByKey(lambda x,y: x + y)
wordCountsSorted = wordCounts.map(lambda pair: (pair[1], pair[0])).sortByKey()
results = wordCountsSorted.collect()
```

```python
for result in results:
    count = str(result[0])
    word = result[1].encode('ascii','ignore').decode()
    if(word):
        print (word +"\t" + count)
```
